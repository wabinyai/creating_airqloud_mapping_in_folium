# creating map using folium

```python
import folium
from folium.plugins import Draw
from folium import plugins
import pandas as pd
```

```python


def create_draw_map(center):
    # Create a folium map centered on the specified coordinates
    m = folium.Map(location=center, zoom_start=12)

    # Add the draw control to the map
    draw = Draw(export=True)
    draw.add_to(m)
    # Add search icone
    plugins.Geocoder().add_to(m)
    # Load the CSV file into a pandas dataframe
    df = pd.read_csv('devices.csv')

    # Loop through each row in the dataframe and add a marker to the map
    for index, row in df.iterrows():
        name = row['name']
        lat = row['latitude']
        lon = row['longitude']
       
        folium.Marker(location=[lat, lon], tooltip=name).add_to(m)

    return m
```

```python
center = [0.3211264,32.5943296]
m = create_draw_map(center)
m.save('map.html')
m
```

```python
import folium
import json

# Load the JSON file with the polygon data
with open('C:/Users/User/Downloads/data.geojson') as f:
    polygon_data = json.load(f)

# Create a Folium map centered on the coordinates of the first polygon
mm = folium.Map(location=[polygon_data['features'][0]['geometry']['coordinates'][0][0][1],
                          polygon_data['features'][0]['geometry']['coordinates'][0][0][0]],
               zoom_start=10)

# Add the polygons to the map
folium.GeoJson(polygon_data).add_to(m)
# Define a style function that returns a dictionary with the desired style properties
def style_function(feature):
    return {
        'fillColor': 'blue',
        'color': 'blue',
        'weight': 1,
        'fillOpacity': 0.2,
    }

# Add the polygons to the map with the style function applied
folium.GeoJson(polygon_data, style_function=style_function).add_to(mm)
# Search “Kampala ” and display the results on the map. 
# “Kampala” is the address of the first polygon in the file.

plugins.Geocoder().add_to(mm)
# Display the map
mm
```
